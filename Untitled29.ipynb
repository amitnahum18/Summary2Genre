# `classify_texts` Function Overview

The `classify_texts` function classifies a list of text strings into predefined categories.

## Inputs
- **texts**: `list[str]` – Texts to classify (English or Hebrew).  
- **labels**: `list[str]` – Categories or genres to compare against.  
- **top_k**: `int` – Number of top categories to return per text (default: 3).

## Process
1. Compute embeddings for each label and each text using a multilingual SentenceTransformer model.  
2. Calculate cosine similarity between each text embedding and label embeddings.  
3. Select the top `top_k` labels for each text based on similarity scores.

## Outputs
Returns a list of dictionaries. Each dictionary contains:  
- `text` – The original text.  
- `top_labels` – Top categories predicted for the text.  
- `scores` – Similarity scores between the text and the predicted categories (0–1 scale).

## Example Usage
```python
texts = ["A young wizard discovers a hidden world of magic.", "ספרות קלאסית המתרחשת בתקופה הרומית."]
labels = ["Science Fiction & Fantasy", "Literature & Classics", "Romance"]
results = classify_texts(texts, labels, top_k=2)
print(results)


````python
from sentence_transformers import SentenceTransformer, util
import torch

model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

default_labels = [
    "Science Fiction & Fantasy", "Literature & Classics", "Romance",
    "Mystery & Thriller", "History & Historical Fiction", "Philosophy & Religion",
    "Science & Popular Science", "Technology & Computer Science", "Psychology & Self-Help",
    "Business & Economics", "Politics & Social Sciences", "Art, Music & Design",
    "Drama & Theatre", "Poetry", "Health & Medicine", "Travel & Geography",
    "Cooking & Lifestyle", "Sports & Games", "Comics, Graphic Novels & Manga",
    "Children & Young Adult", "Horror", "Nature & Environment", "Education & Reference",
    "Non-Fiction", "Other"
]

def classify_texts(texts, labels=default_labels, top_k=3):
    label_embeddings = model.encode(labels, convert_to_tensor=True, normalize_embeddings=True)
    text_embeddings = model.encode(texts, convert_to_tensor=True, normalize_embeddings=True)
    cos_scores = util.cos_sim(text_embeddings, label_embeddings)
    results = {}
    for i, text in enumerate(texts):
        scores = cos_scores[i]
        top_results = scores.topk(top_k)
        selected_labels = [labels[idx] for idx in top_results[1]]
        results[text] = selected_labels
    return results

texts = [
    "הספר עוסק בחברה עתידנית שבה הממשלה שולטת בזיכרונות האזרחים.",
    "A young wizard discovers a hidden world of magic and adventure.",
    "ספרות קלאסית המתרחשת בתקופה הרומית."
]

classified = classify_texts(texts)
for t, cats in classified.items():
    print(f"Text: {t}\nPredicted Categories: {cats}\n")
````

````python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

labels = [
    "Science Fiction & Fantasy", "Literature & Classics", "Romance",
    "Mystery & Thriller", "History & Historical Fiction", "Philosophy & Religion",
    "Science & Popular Science", "Technology & Computer Science", "Psychology & Self-Help",
    "Business & Economics", "Politics & Social Sciences", "Art, Music & Design",
    "Drama & Theatre", "Poetry", "Health & Medicine", "Travel & Geography",
    "Cooking & Lifestyle", "Sports & Games", "Comics, Graphic Novels & Manga",
    "Children & Young Adult", "Horror", "Nature & Environment", "Education & Reference",
    "Non-Fiction", "Other"
]

label_embeddings = model.encode(labels, convert_to_tensor=True, normalize_embeddings=True)

texts = [
    "הספר עוסק בחברה עתידנית שבה הממשלה שולטת בזיכרונות האזרחים.",
    "A young wizard discovers a hidden world of magic and adventure.",
    "ספרות קלאסית המתרחשת בתקופה הרומית."
]

text_embeddings = model.encode(texts, convert_to_tensor=True, normalize_embeddings=True)

cos_scores = util.cos_sim(text_embeddings, label_embeddings)  # [num_texts, num_labels]

top_k = 3
for i, text in enumerate(texts):
    scores = cos_scores[i]
    top_results = scores.topk(top_k)
    selected_labels = [labels[idx] for idx in top_results[1]]
    print(f"Text: {text}\nPredicted Categories: {selected_labels}\n")
````
